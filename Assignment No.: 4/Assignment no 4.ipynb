```python
          #----------------- Assignment no.; 4 ----------------#

''' Group Members: 1) Amol Popalgaht.(251)
                   2) Aditya Bhange. (261)
                   3) Navinya (243)

    Div.: B  ; Batch: B(3)
    Data set: grains.csv
    Link: https://colab.research.google.com/drive/1MUluhGvzB4N7wSSCAo285sQURvm6VgYY#scrollTo=F1J_FkYyltKo&printMode=true '''

import pandas as pd
# Reading the given dataset "Grains.csv"
df = pd.read_csv("/content/sample_data/grains.csv")
print("Output: ")
print(df)
print("")

#<------------------------------------------------------------------------>#
# Q1. Find the no of grains in Andaman and Nicobar which have minimum price greater than 6000?
filtered_data = df[(df['state'] == 'Andaman and Nicobar') & (df['min_price'] > 6000)]

# Count the number of grains
num_grains = len(filtered_data)

# Print the result
print("1) Number of grains with minimum price greater than 6000 in Andaman and Nicobar:", num_grains)

#<------------------------------------------------------------------------>#
# Q2. Which grain get more price in this data set?
highest_price_row = df[df['max_price'] == df['max_price'].max()]
Max_Price = df["max_price"].max()

# Get the commodity of the grain with the highest price
grain_with_highest_price = highest_price_row.iloc[0,3]

# Print the result
print("2) Grain with the highest price:", grain_with_highest_price,'=',Max_Price)

#<------------------------------------------------------------------------>#
# Q3. find the no. of types of grains in this dataset?
filtered_data = df.groupby("commodity").count()

# Count the number of grains
num_grains = len(filtered_data)

# Print the result
print("3) Number of types of grains in this dataset is:", num_grains)

#<------------------------------------------------------------------------>#
# Q4. What is the average modal price across all commodities in the dataset?
average_modal_price = df['modal_price'].mean()

# Print the result
print("4) Average modal price across all commodities:", average_modal_price)

#<------------------------------------------------------------------------>#
# Q5. How many different districts are represented in the dataset?
num_district = df["district"].nunique()

# Print the result
print("5) Number of different districts are represented in the dataset is:",num_district)

#<------------------------------------------------------------------------>#
# Q6. What is the most common variety of rice in the dataset?
rice_data = df[df["commodity"]=="Rice"]

# most_common_rice_variety
most_common_rice_variety = rice_data['variety'].mode().values[0]

# Print the result
print("6) The most common variety of rice in the dataset is:",most_common_rice_variety)

#<------------------------------------------------------------------------>#
# Q7. How many commodities have a modal price greater than 10,000?
filtered_data = df[df['modal_price'] > 10000]

# Count the number of commodities
num_commodities = len(filtered_data)

# Print the result
print("7) Number of commodities with a modal price greater than 10,000:", num_commodities)

#<------------------------------------------------------------------------>#
# Q8. Which district has the lowest minimum price for potatoes?
potato_data = df[df['commodity'] == 'Potato']

#  district with the lowest minimum price
district_lowest_price = potato_data[potato_data['min_price'] == potato_data['min_price'].min()]['district'].values[0]

# Print the result
print("8) The district has the lowest minimum price for potatoes is:", district_lowest_price)

#<------------------------------------------------------------------------>#
# Q9. What is the average maximum price for bananas across all districts
Bananas_data = df[df['commodity'] == 'Banana']

# finding average_max_price for bananas
average_max_price = df['max_price'].mean()

# Print the result
print("9) Average maximum price for bananas across all districts is:", average_max_price)

#<------------------------------------------------------------------------>#
# Q10. Which state has the highest number of commodities listed in the dataset?
state_commodity_counts = df['state'].value_counts()

# Finding the state with the highest number of commodities
state_with_highest_commodities = state_commodity_counts.idxmax()

# Print the result
print("10) State with the highest number of commodities listed:", state_with_highest_commodities)

#<------------------------------------------------------------------------>#
# Q11. What is the modal price for Jute in the market "Islampur"?
filtered_data = df[(df['commodity'] == 'Jute') & (df['market'] == 'Islampur')]

# Retrieve the modal price for Jutein Islampur
modal_price_Jute_islampur = filtered_data['modal_price'].values[0]

# Print the result
print("11) Modal price for Jute in Islampur is:", modal_price_Jute_islampur)

#<------------------------------------------------------------------------>#
# Q12. How many commodities have a minimum price between 5,000 and 7,000?
filtered_data = df[(df['min_price'] >= 5000) & (df['min_price'] <= 7000)]

# Count the number of commodities
num_commodities = len(filtered_data)

# Print the result
print("12) Number of commodities with a minimum price between 5,000 and 7,000 is:", num_commodities)

#<------------------------------------------------------------------------>#
# Q13. How many commodities have a modal price greater than both the minimum and maximum prices?
filtered_data = df[df['modal_price'] > df[['min_price', 'max_price']].max(axis=1)]

# Count the number of commodities
num_commodities = len(filtered_data)

# Print the result
print("13) Number of commodities with a modal price greater than both the minimum and maximum prices is:", num_commodities)

#<------------------------------------------------------------------------>#
# Q14. How many markets are there in the state "West Bengal"?
filtered_data = df[df['state'] == 'West Bengal']

# Count the number of unique markets
num_markets = filtered_data['market'].nunique()

# Print the result
print("14) Number of markets in the state West Bengal is:", num_markets)

#<------------------------------------------------------------------------>#
# Q15. What is the variety of potato with the highest modal price?
potato_data = df[df['commodity'] == 'Potato']

# Find the variety of potato with the highest modal price
variety_highest_modal_price = potato_data[potato_data['modal_price'] == potato_data['modal_price'].max()]['variety'].values[0]

# Print the result
print("15) Variety of potato with the highest modal price is:", variety_highest_modal_price)

#<------------------------------------------------------------------------>#
# Q16. Which type of commodity found in Chittor district?
commodity = df[df['district']=='Chittor']['commodity'].unique()

# Print the result
print("16) Type of commodity found in Chittor district is:",commodity)

#<------------------------------------------------------------------------>#
# Q17. How Many Varities are "Local"?
local = len(df[df['variety']=='Local'])

# Print the result
print("17) Number of Local varitirs are:",local)

#<------------------------------------------------------------------------>#
# Q18. Find the sum of Modal price of "carrot" and "Onion"?
Total_modal_price = df[df['commodity'].isin(['Carrot','Onion'])]['modal_price'].sum()

# Print the result
print("18) sum of Modal price of carrot and Onion is:",Total_modal_price)

#<------------------------------------------------------------------------>#
# Q19. Find the average max price of Cauliflower?
average_max_price = df[df['commodity']=='Cauliflower']['max_price'].mean()

# Print the result
print("19) The average max price of Cauliflower is:",average_max_price)

#<------------------------------------------------------------------------>#
# Q20. Find no. of district which are also markets?
num_districts_markets = df[df['district'].isin(df['market'].unique())]['district'].nunique()

# Print the result
print("20) Number of districts that are also markets:", num_districts_markets)
```
